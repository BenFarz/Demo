```python
import tensorflow as tf
```

```python
X=tf.expand_dims(tf.cast(tf.range(-100,100,3), dtype=tf.float16), axis=-1)
```

```python
tf.shape(X)
```

```python
tf.size(X)
```

```python
X_train = X[:45]
X_test = X[45:]
```

```python
y=X+2.5
```

```python
y_train = y[:45]
y_test = y[45:]
```

```python
tf.random.set_seed(43)
mdl1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
mdl1.compile(loss= tf.keras.losses.mae,
             optimizer = tf.keras.optimizers.SGD(),
             metrics = ["mae"])
mdl1.fit(X_train, y_train, epochs = 100)
```

```python
tf.random.set_seed(43)
mdl2 = tf.keras.Sequential([
tf.keras.layers.Dense(20),
tf.keras.layers.Dense(1, input_shape=[1])
])
mdl2.compile(loss=tf.keras.losses.mae,
             optimizer = tf.keras.optimizers.SGD(),
             metrics = ["mae"])
mdl2.fit(X_train, y_train, epochs=100)
```

```python
mdl3=tf.keras.Sequential([
    tf.keras.layers.Dense(20),
    tf.keras.layers.Dense(1, input_shape=[1])
])
mdl3.compile(loss=tf.keras.losses.mae,
             optimizer= tf.keras.optimizers.Adam(lr=.1),
             metrics=["mae"])
mdl3.fit(X_train, y_train, epochs=300)
```

```python
tf.random.set_seed(43)
mdl4 = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1, input_shape = [1])
])
mdl4.compile(loss = tf.keras.losses.mae,
             optimizer = tf.keras.optimizers.SGD(),
             metrics = ["mae"])
mdl4.fit(X_train, y_train, epochs = 300)
```

```python
y_pred1= mdl1.predict(X_test)
y_pred2= mdl2.predict(X_test)
y_pred3=mdl3.predict(X_test)
y_pred4=mdl4.predict(X_test)
```

```python
import matplotlib.pyplot as plt
def draw(in_data = X_test,
         o=y_test,
         o1=y_pred1,
         o2=y_pred2,
         o3=y_pred3,
         o4=y_pred4):
  plt.figure(figsize = (10,7))
  plt.scatter(in_data, o, c="g", label="BenchMark")
  plt.scatter(in_data, o1, c="r", label="m1")
  plt.scatter(in_data, o2, c="b", label="m2")
  plt.scatter(in_data, o3, c="y", label="m3")
  plt.scatter(in_data, o4, c="m", label="m4")
  plt.legend();
```

```python
draw()
```
